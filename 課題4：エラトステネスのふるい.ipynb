```python
prime = []
for i in range(1, 101):
    if i == 1:
        continue
    elif i != 2 and i % 2 == 0:
        continue
    elif i != 3 and i % 3 == 0:
        continue
    elif i != 5 and i % 5 == 0:
        continue
    elif i != 7 and i % 7 == 0:
        continue
    else:
        prime.append(i)

print(prime)

```
