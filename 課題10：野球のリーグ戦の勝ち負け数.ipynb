```python
class Baseball_Team:
    name = ""
    win = 0
    lose = 0
    draw = 0

    def __init__(self, name, win, lose, draw):
        self.name = name
        self.win = win
        self.lose = lose
        self.draw = draw
    
    def calc_win_rate(self):
        if self.win + self.lose == 0:
            return 0
        else:
            return round(self.win / (self.win + self.lose), 3)
    
    def show_team_result(self):
        rate = self.calc_win_rate()
        print(f"{self.name.ljust(8)}{str(self.win).rjust(5)}{str(self.lose).rjust(5)}{str(self.draw).rjust(5)}{str(rate).rjust(6)}")
    
    @classmethod
    def show_header(self):
        print(f"{"team".ljust(8)}{"win".rjust(5)}{"lose".rjust(5)}{"draw".rjust(5)}{"rate".rjust(6)}")
```

```python
results = [
    ["Archers", 77, 64, 2],
    ["Meteors", 71, 69, 3],
    ["Cheetahs", 69, 68, 6],
    ["Sharks", 70, 70, 3],
    ["Pythons", 68, 73, 2],
    ["Owls", 59, 82, 2]
]

Baseball_Team.show_header()
for result in results:
    Baseball_Team(*result).show_team_result()
```
