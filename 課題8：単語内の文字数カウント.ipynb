```python
import string
import re

# キーがアルファベット小文字で、値が 0 の辞書を作成する
keys = list(string.ascii_lowercase)
alphabet = dict.fromkeys(keys, 0)

# 入力された英単語
words = []

while True:
    word = input("英単語を入力してください: ")

    # 入力された文字が "" なら終了
    if word == "":
        break

    # a-z のみ入力可
    if not re.match('^[a-z]+$', word):
        print("エラー: 英単語は a から z のアルファベットを小文字で入力してください")
        continue
    
    # 英単語を追加
    words.append(word)
    
    # 入力された英単語に踏まれるアルファベットをカウント
    for c in word:
        alphabet[c] += 1

# 英単語一覧をソート(昇順)して表示
print(f"入力した英単語: {sorted(words)}")

# 1文字以上のアルファベットを表示
for key in alphabet:
    if alphabet[key] > 0:
        print(f"{key}が{alphabet[key]}個ありました")
```
